```python
df=df[df.Date>="2017-10-01"]
df.head()
```

```python
print(df["temp(F)"].dtype)
```

```python
df['temp(F)'] = df['temp(F)'].str.rstrip('F')
df["temp(F)"].head()
```

```python
df['dewpoint(F)'] = df['dewpoint(F)'].str.rstrip('F')
df["dewpoint(F)"].head()
```

```python
df['humidity(%)'] = df['humidity(%)'].str.rstrip('%')
df["humidity(%)"].head()
```

```python
df['Windspeed(mph)'] = df['Windspeed(mph)'].str.rstrip('mph')
df["Windspeed(mph)"].head()
```

```python
df['Precip(in)'] = df['Precip(in)'].str.rstrip('in')
df["Precip(in)"].head()
```

```python
df[["temp(F)","dewpoint(F)","humidity(%)","Windspeed(mph)","Precip(in)"]] = df[["temp(F)","dewpoint(F)","humidity(%)","Windspeed(mph)","Precip(in)"]].astype(float)
print(df["temp(F)"].dtype)
```

```python
# Monthly Mean Temperature Trend
df["Date"]=pd.to_datetime(df["Date"])
df.resample('M', on = 'Date')["temp(F)"].mean().plot()
plt.title("Monthly Mean Temperature Trend")
plt.show()
```

```python
# Daily Mean Temperature Trend
df.resample('D', on = 'Date')["temp(F)"].mean().plot()
plt.title("Daily Mean Temperature Trend")
plt.show()
```

```python
# average trip duration based on each month 
df1.resample('M', on = 'Start date')['Duration seconds'].mean().plot()
plt.title("average trip duration based on each month in seconds")
plt.show()
```
